## 1. 컨볼루셔널 네트워크
- CNN(Convolutional neural network)
- 전통적인 뉴럴 네트워크 앞에 컨볼루셔널 계층을 붙인 형태
- 컨볼루셔널 계층을 통해서 입력 받은 이미지에 대한 특징(Feature)를 추출하게 되고, 추출한 특징을 기반으로 기존의 뉴럴 네트워크를 이용하여 분류

## 2. 컨볼루셔널 레이어(Convolutional Layer)
- 입력 데이터로 부터 특징을 추출하는 역할
- 특징을 추출하는 필터(Filter)를 사용함
- 필터의 값을 비선형 값으로 바꿔주는 액티베이션(Activation)함수를 사용

### 2-1. 필터(Filter)
- 특징이 데이터에 있는지 없는지 검출하는 함수
- 필터는 구현에서 행렬로 정의
- 입력 받는 이미지 모두 행렬로 변환
- 입력받은 데이터에서 그 특징을 가지고 있으면 결과값이 큰 값이 나오며, 특징을 가지고 있지 않으면 0에 가까운 값이 변환

필터 - 행렬로 정의



- 그림이 있는 부분만 값이 나옴
- 나머지는 0 임

### 2-2. Stride
- 필터를 적용하는 간격(예: 우측으로 한칸씩 아래로 한칸씩 적용) 값
- 필터를 적용해서 얻어낸 결과를 Feature map 또는 activation map라고 부름



            3*3 행렬이 한칸씩 움직여 비교        각 행렬곱을 통해 결과값 추출
            총 9번 움직여 비교

### 2-3. padding
- CNN 네트워크 특징 중 여러 단계에 걸쳐서 계속 필터를 연속적으로 적용하는데 필터 적용 후 결과 값이 작아지게 되면 처음에 비해 특징이 많이 유실될 수 있음
- 충분히 특징이 추출되기 전에 결과 값이 작이지면 특징이 유실되므로 이를 방지하기 위해 padding 기법을 사용
- 입력값 주위로 0값을 넣어서 입력 값의 크기를 인위적으로 키워 결과값이 작이지는 것을 방지
- 원본 데이터에 0값을 넣어서 원래의 특징을 희석 시키고 이것을 기반으로 머신러닝 모델이 트레이닝 값에만 정확하게 맞아 들어가는 오버피팅도 방지

### 2-4. activation function
- 필터를 통해서 추출한 값이 들어가 있는 큰 값고 0에 가까운 값들을 비선형 값으로 바꿔주는 과정이 필요함
- sigmoid → relu
    - ↑ back propagation함수가 잘 작동하지 않음

### 2-5. pooling
- 작은 사진을 가지고도 그 사진의 내용이 어떤 사진인지 판단할 수 있는 원리를 이용한 기법
- max polling(가장 많이 사용), average pooling, L2-norm pooling
- 엑티베이션 함수 마다 매번 적용하는 것이 아니라 데이터의 크기를 줄이고 싶을 때 선택적으로 사용

### 2-6. max pooling
- Activation map을 M*N의 크기를 잘라낸 후 그 안에서 가장 큰 값을 뽑아내는 방법
- 전체 데이터의 사이즈가 줄어들기 때문에 연산에 들어가는 리소스가 적어짐
- 데이터의 크기를 줄이면서 소실이 발생하기 때문에 오버피팅을 방지할 수 있음

### 2-7. Fully connected Layer
- 기존의 뉴럴 네트워크

### 2-8. Dropout 계층
- 오버피팅을 막기 위한 방법
- 뉴럴 네트워크가 학습 중일 때 랜덤하게 뉴럴을 꺼내서 학습을 방해함으로 학습이 학습용 데이터에 치우치는 것을 막아줌
