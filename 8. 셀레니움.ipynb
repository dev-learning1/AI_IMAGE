### 네이트 실시간 검색어
* https://www.nate.com/

```python
driver = webdriver.Chrome('D:/Jupyter Project/data_ai/chromedriver.exe')  
driver.get('https://www.nate.com/')
```

```python
from bs4 import BeautifulSoup
```

```python
html = driver.page_source  # html 몽땅 가져오기
soup = BeautifulSoup(html)

issues = soup.findAll('a', {'class':'ik'})
print(issues)
```

```python
'''
실시간 이슈         상승/하락
--------------------------------
집무실 이전 윤      - 
조재진              - 
골 때리는 그녀들   ↑2
유 퀴즈 온 더 블럭  -
한국 유럽으로 LNG  ↓2
'''
```

```python
for issue in issues:
    print(issue.text.strip().split('\n')[0], end=' ')
    if issue.text.strip().split('\n')[1] == '동일':
        print('-')
    elif '상승' in issue.text.strip().split('\n')[1]:
        print('↑' + issue.text.strip().split('\n')[1][-1])
    elif '하락' in issue.text.strip().split('\n')[1]:
        print('↓' + issue.text.strip().split('\n')[1][-1])
    elif issue.text.strip().split('\n')[1] == 'new':
        print('new')
```

```python
# 페이징
# 위의 코드를 차례대로 해야함 (크롬창 열고 -> 배스트댓글 -> 전체댓글)
# 클릭할 때마다 src주소에서 뒤에 []안에 숫자가 바뀜 (2~다음 페이지까지만 해당됨)(12번부터는 숫자 고정)
# 크롤링 못하게 하려고 일부러 이렇게 만들었나...?

# 2번 XPath : //*[@id="cbox_module"]/div/div[7]/div/a[1]
# 3번 XPath : //*[@id="cbox_module"]/div/div[7]/div/a[3]
# 4번 XPath : //*[@id="cbox_module"]/div/div[7]/div/a[4]
# 5번 : //*[@id="cbox_module"]/div/div[7]/div/a[5]
# 6번 : //*[@id="cbox_module"]/div/div[7]/div/a[6]
# 10번 : //*[@id="cbox_module"]/div/div[7]/div/a[10]
# 다음 : //*[@id="cbox_module"]/div/div[7]/div/a[11]

# 11번 : 현재 페이지에서 댓글 크롤링하기
# 12번 : //*[@id="cbox_module"]/div/div[7]/div/a[3]
# 13번 : //*[@id="cbox_module"]/div/div[7]/div/a[4]
# 14번 : //*[@id="cbox_module"]/div/div[7]/div/a[5]
# 18번 : //*[@id="cbox_module"]/div/div[7]/div/a[9]
# 20번 : //*[@id="cbox_module"]/div/div[7]/div/a[11]
# 다음 : //*[@id="cbox_module"]/div/div[7]/div/a[12]

# 21번 : 현재 페이지에서 댓글 크롤링하기
# 22번 : //*[@id="cbox_module"]/div/div[7]/div/a[3]
# 23번 : //*[@id="cbox_module"]/div/div[7]/div/a[4]
# 나머지 똑같음


# 2번 마지막번호가 1이기 때문에 자동화(for문)에 넣을 수 없음
import time

print('***** 전체 댓글2 *****')
driver.find_element_by_xpath('//*[@id="cbox_module"]/div/div[7]/div/a[1]').click()
time.sleep(1)

html = driver.page_source
soup = BeautifulSoup(html)

comment_area = soup.findAll('span', {'class','u_cbox_contents'})

for i in range(len(comment_area)):
        comment = comment_area[i].text.strip()
        print(comment)
        print('-'*80)

# 자동화 (3페이지 ~ 10페이지)     
for i in range(3,11):
    #xpath = '//*[@id="cbox_module"]/div/div[7]/div/a[{}]'.format(i)
    driver.find_element_by_xpath('//*[@id="cbox_module"]/div/div[7]/div/a['+str(i)+']').click()
    time.sleep(1)

    html = driver.page_source
    soup = BeautifulSoup(html)

    comment_area = soup.findAll('span', {'class','u_cbox_contents'})
    # print(comment_area)

    print('***** 전체 댓글{} *****'.format(i))
    for i in range(len(comment_area)):
        comment = comment_area[i].text.strip()
        print(comment)
        print('-'*80)
        
# '다음'으로 넘어가서 11~20페이지 읽고 다시 '다음'으로 넘어가서 읽는 것을 반복
```
