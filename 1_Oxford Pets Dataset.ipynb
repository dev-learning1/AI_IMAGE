```python
!pip install opencv-python
import cv2
import xml.etree.ElementTree as et
from matplotlib.patches import Rectangle
```

```python
# 같은 이미지를 같은 경로로 설정
images_path = image_files[3]
print(images_path)
bbox_path = images_path.replace(image_dir, bbox_dir).replace('jpg', 'xml')
print(bbox_path)
```

```python
image = cv2.imread(images_path)
print(image)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# 파싱
tree = et.parse(bbox_path)
tree
```

```python
xmin = float(tree.find('./object/bndbox/xmin').text)
xmax = float(tree.find('./object/bndbox/xmax').text)
ymin = float(tree.find('./object/bndbox/ymin').text)
ymax = float(tree.find('./object/bndbox/ymax').text)
print(xmin, xmax, ymin, ymax)

rect_x = xmin
rect_y = ymin
rect_w = xmax - xmin
rect_h = ymax - ymin
```

```python
# Rectangle(이미지, width, height, filll, color)
# Rectangle((시작점 좌표), (종료점 좌표), fill, color)
# Rectangle((시작점 좌표), width, height, fill, color)

rect = Rectangle((rect_x, rect_y), rect_w, rect_h, fill=False, color='red')
plt.axes().add_patch(rect)
plt.imshow(image)
plt.show()
```

```python
image_path = image_files[30]
print(image_path)
seg_path = image_path.replace(image_dir, seg_dir).replace('jpg', 'png')
print(seg_path)

image = cv2.imread(image_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
seg_map = cv2.imread(seg_path, cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(15, 15))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.subplot(1, 2, 2)
plt.imshow(seg_map)

plt.show()
```

### 교차검증
- 데이터의 수가 적은 경우 데이터 중의 일부인 검증 데이터의 수도 적기 때문에 검증 성능의 신뢰도가 떨어짐
- 검증 데이터의 수를 증가시키면 학습용 데이터의 수가 적어지므로 정상적인 학습이 되지 않음
- 데이터 수가 적은 경우 해결하기 위한 검증 방법
- 과적함을 해결하기 위한 방법

```python
# !pip install sklearn
from sklearn.model_selection import KFold
```

```python
df
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=40)
df['fold'] = -1# -1은 의미없음. 단순히 초기값
for idx, (t, v) in enumerate(kf.split(df), 1):
    print(t, v, len(v))
    df.loc[v, 'fold'] = idx     #df.loc[row, col]
```

```python
df
```

```python
print(len(df[df['fold'] == 1]))
```

```python
print(len(df[df['fold'] != 1]))
```

```python
value_counts = df[df['fold'] != 5]['id'].value_counts().sort_index()
plt.bar(range(len(value_counts)), value_counts.values, align='center')
plt.xticks(range(len(value_counts)), value_counts.index.values)
plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=40)
# test, train 비율 정하기
#skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=40, train_size=5, test_size=5)

df['fold'] = -1# -1은 의미없음. 단순히 초기값
for idx, (t, v) in enumerate(skf.split(df, df['id']), 1):
    print(t, v, len(v))
    df.loc[v, 'fold'] = idx     #df.loc[row, col]
```

```python
value_counts = df[df['fold'] != 5]['id'].value_counts().sort_index()

plt.bar(range(len(value_counts)), value_counts.values, align='center')
plt.xticks(range(len(value_counts)), value_counts.index.values)
plt.tight_layout()
plt.show()
```
