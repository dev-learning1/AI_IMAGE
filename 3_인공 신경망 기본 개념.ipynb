### Perceptron

- 하나의 선을 기준으로 분류(Perceptron)






## 2. MLP(Multi Layer Perceptron) 다중 계층 퍼셉트론

- 하나의 선으로 구분 안됨


- MLP도 레이어가 복잡해 질 수록 연산이 복잡해져서 w와 b의 값을 구하는 것이 불가능할 수 있음
- 위 문제를 해결하기 위해 Back propagation이라는 알고리즘이 도입(뉴럴 네트워크를 순방향으로 한번 연산한 뒤, 그 결과를 가지고 뉴럴 네트워크를 역방향으로 계산하면서 값을 구한다는 개념)
- Back propagation 또한 뉴럴 네트워크의 깊이가 깊어질수록 Back propagation이 제대로 동작하지 않는 문제점이 생김(순방향 결과를 역방향에 반영하면서 계산해야 하는데 레이어가 깊어질수록 뒤에 있는 값이 앞으로 전달되지 않는 문제 -> Vanishing Gradient 문제)
- ReLu라는 activation function으로 위 문제를 해결
